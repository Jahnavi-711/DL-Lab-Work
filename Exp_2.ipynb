```python
# Experiment 2

import numpy as np
from sklearn.linear_model import SGDRegressor

celsius = np.array([-40, -10, 0, 8, 15, 22, 38]).reshape(-1, 1)
fahrenheit = np.array([-40, 14, 32, 46, 59, 72, 100])

model = SGDRegressor(max_iter=1000, learning_rate='constant', eta0=0.001)

model.fit(celsius, fahrenheit)

user_c = float(input("Enter temperature in Celsius: "))

predicted_f = model.predict([[user_c]])[0]

actual_f = (user_c * 9/5) + 32

print(f"Predicted Fahrenheit (Model): {round(predicted_f, 2)}")
print(f"Actual Fahrenheit (Formula): {round(actual_f, 2)}")
```
