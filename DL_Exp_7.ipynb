```python
import os
import shutil
import random

# Remove old dataset if exists
if os.path.exists("dataset_split"):
    shutil.rmtree("dataset_split")

os.makedirs("dataset_split/train/cats", exist_ok=True)
os.makedirs("dataset_split/train/dogs", exist_ok=True)
os.makedirs("dataset_split/test/cats", exist_ok=True)
os.makedirs("dataset_split/test/dogs", exist_ok=True)

def split_data(source, train, test, split=0.8):
    files = os.listdir(source)
    files = [f for f in files if not f.startswith('.')]
    random.shuffle(files)

    split_size = int(len(files) * split)

    for file in files[:split_size]:
        shutil.copy(os.path.join(source, file), train)

    for file in files[split_size:]:
        shutil.copy(os.path.join(source, file), test)

split_data("cats_set", "dataset_split/train/cats", "dataset_split/test/cats")
split_data("dogs_set", "dataset_split/train/dogs", "dataset_split/test/dogs")

print("Split complete!")
```

```python
print("Train cats:", len(os.listdir("dataset_split/train/cats")))
print("Train dogs:", len(os.listdir("dataset_split/train/dogs")))
print("Test cats:", len(os.listdir("dataset_split/test/cats")))
print("Test dogs:", len(os.listdir("dataset_split/test/dogs")))
```

```python
import tensorflow as tf

train_dir = "dataset_split/train"
test_dir = "dataset_split/test"

train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    train_dir,
    image_size=(100, 100),
    batch_size=16
)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    test_dir,
    image_size=(100, 100),
    batch_size=16
)

AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.map(lambda x, y: (x/255.0, y)).cache().prefetch(AUTOTUNE)
test_ds = test_ds.map(lambda x, y: (x/255.0, y)).cache().prefetch(AUTOTUNE)
```
